```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\bhara\Downloads\CSV\ML\covid.csv")
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
plt.plot(df['Date'], df['Confirmed'], marker='o')
plt.title("COVID-19 Confirmed Cases Trend")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.show()
```

```python
df['Death_Rate'] = (df['Deaths'] / df['Confirmed']) * 100
df[['Date', 'Death_Rate']]
```

```python
df['Death_Rate'].plot(kind='bar')
plt.title("Death Rate")
plt.xlabel("Date")
plt.ylabel("Death_rate")
plt.show()
```

```python
plt.plot(df['Date'], df['Recovered'], color='green')
plt.title("Recovery Trend")
plt.xlabel("Date")
plt.ylabel("Recovered Cases")
plt.show()
```
