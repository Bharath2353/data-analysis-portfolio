```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\bhara\Downloads\CSV\ML\stock.csv")
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
plt.plot(df['Date'], df['Close'], marker='o')
plt.title("Stock Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Price (â‚¹)")
plt.show()
```

```python
df['Daily_Return'] = df['Close'].pct_change() * 100
df[['Date', 'Daily_Return']]
```

```python
df['Daily_Return'].plot(kind='bar')
plt.title("Daily_Return")
plt.xlabel("Date")
plt.ylabel("Daily_Rerurn")
plt.show()
```

```python
plt.bar(df['Date'], df['Volume'])
plt.title("Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```
